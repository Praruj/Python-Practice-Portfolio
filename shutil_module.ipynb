# Shutil Module: copy,move, rename and Delete

`Shutil`: Short for shell utilities. Shell is another term for terminal command line. Windows uses (\), whereas, (/) works on all operating systems. 

```python
from pathlib import Path
h=Path.home()
(h/"Spam").mkdir(exist_ok=True)
with open(h/"Spam/file1.txt","w",encoding="utf-8") as file:
    file.write("hello! honey?")
```

```python
from pathlib import Path
import shutil

h=Path.home()
shutil.copy(h/"spam/file1.txt",h)
shutil.copy(h/"spam/file1.txt", h/"spam/file2.txt")

# shutil.copy(s,d) copies single file only  
```

```python
import shutil
from pathlib import Path

h=Path.home()
shutil.copytree(h/"spam",h/"spam_backup")

#shutil.copytree() creates a new folder name spam_backup with same content as spam folder(original). It copies folder at path source, along with all of its files. 
```

```python
# moving and Renaming the folders at the Path source to path destination and return a string of a new location`s  absolute path.

import shutil
from pathlib import Path
h=Path.home()
(h/"spam2").mkdir(exist_ok=True)
shutil.move(h/"spam/file1.txt" , h/"spam2")
```

```python
# if the destination path is not an existing folder, shutil.move() will use this path to rename the file. 
shutil.move(h/"spam/file2.txt", h/"spam2/new_name.txt")
```

## Permanently Deleting files and Folders

We can delete single file or single empty folder with fuinctions in the os module, whereas to delete a folder and all of its contents:

- Calling `shutil.rmtree(Path)` will delete the entire folder tree at path, including all files and subfolders it contains.
- Calling `os.unlink(Path)` will delete the single file at the Path
- Calling `os.rmdir(path)` will delete the folder at path. This folder must be empty.
