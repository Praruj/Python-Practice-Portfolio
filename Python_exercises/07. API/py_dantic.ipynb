# The Pydantic library

```python
from pydantic import BaseModel,EmailStr,Field
from typing import List, Optional, Dict, Annotated


class Showcase(BaseModel):

    name:str=Field(max_length=50)
    model:str
    category:str
    price:float
    status: bool
    version: list[str]                                      #(we can use Dict[str,str] like this as well)
    device: Optional[Dict[str,str]]=None
    email: Optional[EmailStr]=None

def showcase_info(showcase:Showcase):   # used defined pydantic model into function signature
    print(showcase.name)
    print(showcase.model)
    print(showcase.version)
    print("inserted")

info={"name": "Refrigerator Unit",
    "model": "UR-100",
    "category": "Cooling",
    "price": 25000,
    "status": True,
    "version":["1.0","2.0"]}

showcase_obj = Showcase(**info)
showcase_info(showcase_obj)
```
