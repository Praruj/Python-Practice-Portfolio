Making To do list (OOP):
Create `two classes:`

*Task*

- Stores a task description
- Tracks completed status
- Has a method to mark the task as completed

*ToDoList*

- Stores a list of Task objects
- Has methods to add, show, complete, and delete tasks

```python
class Task:
    def __init__(self, description):
        self.description = description
        self.completed = False

    def mark_completed(self):
        self.completed = True


class ToDoList:
    def __init__(self):
        self.tasks = []
    
    def add_task(self, description):
        task = Task(description)
        self.tasks.append(task)
        print("Task added!")

    def show_tasks(self):
        if not self.tasks:
            print("No tasks yet")
        else:
            for i, task in enumerate(self.tasks, 1):
                status = "yes" if task.completed else "no"
                print(f"{i}. [{status}] {task.description}")

    def complete_task(self, number):
        if 1 <= number <= len(self.tasks):
            self.tasks[number - 1].mark_completed()
            print("Task completed!")
        else:
            print("Invalid task number")

    def delete_task(self, number):
        if 1 <= number <= len(self.tasks):
            removed = self.tasks.pop(number - 1)
            print(f"Deleted: {removed.description}")
        else:
            print("Invalid task number")
```

```python
def main():
    todo = ToDoList()

    while True:
        print("\n-- OOP To-Do List --")
        print("1. Add task")
        print("2. Show tasks")
        print("3. Complete task")
        print("4. Delete task")
        print("5. Exit")

        choice = input("Choose: ")

        if choice == "1":
            todo.add_task(input("Task: "))

        elif choice == "2":
            todo.show_tasks()

        elif choice == "3":
            todo.show_tasks()
            todo.complete_task(int(input("Number: ")))

        elif choice == "4":
            todo.show_tasks()
            todo.delete_task(int(input("Number: ")))

        elif choice == "5":
            print("Goodbye!")
            break

        else:
            print("Invalid choice")
            

# Run the program
main()
```
