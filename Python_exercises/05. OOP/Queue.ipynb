# **Queue**

A queue is linear data structure that operates on principle FIFO (first in, first out). Elements are processed in the order they were added. 

*Operations*
- Enqueue : add an element to the rear of the queue.
- Dequeue: removes and returns the elements at the front of queue without removing it.

```python
class Queue: 

    def __init__(self):
        self.items=[]
    

    def enqueue(self,item):
        self.items.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0) 
        else:
            return ("cannot dequeue from an empty queue")


    def is_empty(self):
        return len(self.items)==0

queue=Queue()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
queue.enqueue(40)

print("current items:", queue.items)

dequeued_items=queue.dequeue()
print("dequeued items", dequeued_items)
```
