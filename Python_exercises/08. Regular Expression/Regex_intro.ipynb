## *More Regex*

- `^` : matches the start of string
- `$` : matches the end of string or just before the newline at the end of string
- `[]`: set of characters
- `[^]`: complementing the set


#### Explanation of  `[^@]+`

- `[...]` = character set  
- `^` inside brackets = NOT  
- `[^@]` = any character except `@`  
- `+` = one or more of those characters

`````python
# To make it more robust we can transform this code into:

import re

mail=input("write your email").strip()

if re.search(r"^[a-zA-Z0-9[^@]]+@gmail\.com$",mail):
    print("valid")
else:
    print("invalid")
`````

## Python Regex Flags (`re` module) — Complete Guide

In Python’s `re` module, **flags** are optional settings that change how regular expressions behave.

You pass them like this:

```python
re.search(pattern, text, flags=...)
```
1. `flags=0` means **no special flags are applied**.
- It is the **default behavior** of the `re` module.

2. `flags=1` is an old-style way of enabling **`re.IGNORECASE`**.
- It means the same as:
  ```python id="f1"
  re.IGNORECASE

## Regex Flags in Python (`re` module)

- `re.IGNORECASE`  
  Makes the pattern match letters regardless of case (uppercase or lowercase).

- `re.MULTILINE`  
  Allows `^` and `$` to match the start and end of each line, not just the whole string.

- `re.DOTALL`  
  Makes the dot (`.`) match all characters, including newline (`\n`).
```


`````python
import re

mail=input("write your email").strip()

if re.search(r"^\w+(\.\w+)?@\w+(\.\w+)?\.com$",mail, re.IGNORECASE):
    print("valid")
else:
    print("invalid")
`````

```Python
if re.search(r"^\w+(\.\w+)?@\w+(\.\w+)?\.com$",mail, re.IGNORECASE):
```

Matches an email address where the username and domain consist of word characters optionally separated by a single dot, ending specifically with `.com`, in a case-insensitive way.

## Another Exercise

`re.search()` returns an object, and .group() is a method inside that object.

`````python
import re

name=input("what is your name?").strip()
matches= re.search(r"^(.+),(.+)$",name)

if matches:
    last, first=matches.groups()
    name=f"{first} {last}"

print(f"hello {name}")
`````

OR

`````python
import re

name=input("what is your name?").strip()


if matches:= re.search(r"^(.+), ?(.+)$",name):
    # last=matches.group(1)
    # first=matches.group(2)
    name=matches.group(2)+" "+matches.group(1)
    # name=f"{first} {last}"
print(f"hello, {name}")
`````

## Walrus Operator (`:=`) — Quick Takeaway

The walrus operator (`:=`) in Python lets to  **assign a value and use it in the same expression**.

### Key idea:
> It combines assignment + usage in one step.

### Example:
```python

n=len("hello")
if n>3:
    print(n)


# With Walrus

if (n := len("hello")) > 3:
    print(n)

**On using Regular expression**

`re.sub()` 

### Syntax
```python
re.sub(pattern, replacement, string, count=0, flags=0)

```
### **Parameters**

| Parameter | Description |
| :--- | :--- |
| **pattern** | The regex pattern to search for. |
| **replacement** | The text to replace matches with. |
| **string** | The original string to be searched. |
| **count** | *Optional.* Number of occurrences to replace. Default is 0 (all). |
| **flags** | *Optional.* Regex modifiers (e.g., `re.IGNORECASE`). |

`re.sub()` replaces parts of a string that match a regex pattern with new text.

`````python
import re
url=input("URL:").strip()
username=re.sub(r"^(https?://)?(www\.)?facebook\.com/","",url)   
print(f"Username:{username}")
`````

##### On Using 

`re.search()`:

`re.search()` is used to search a string for the first occurrence of a regex pattern.



## Syntax


```python
re.search(pattern, string, flags=0)
```

pattern  → regex pattern to search for  
string   → original string to search in  
flags    → optional regex flags (`re.IGNORECASE`, etc.)

Returns a Match object if a match is found,  
otherwise returns `None`.

`````python
import re
url=input("URL:").strip()


if matches:= re.search(r"^(?:https?://)?(?:www\.)?facebook\.com/(.+)$",url,re.IGNORECASE):
    print(f"Username:",matches.group(1))
`````
