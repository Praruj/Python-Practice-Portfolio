# OpenPyxl

## What is openpyxl?

`openpyxl` is a Python library used to read, write, and modify Excel files.

## Supported File Formats

`openpyxl` provides native support for Excel 2010+ formats:

- `.xlsx` — Excel workbook
- `.xlsm` — Workbook with macros
- `.xltx` — Excel template
- `.xltm` — Template with macros

## Advantages

- Does **not** require Microsoft Excel to be installed
- Useful for:
  - Server environments
  - Automated reports
  - Data pipelines
- Used by libraries like `pandas` for Excel operations



1. To use this library we need to install:

    ```python
    pip install openpyxl
    ```

2. To use openpyxl in Python scripts, import the necessary classes:

    ```python
    from openpyxl import Workbook, load_workbook
    ```

## Key Concepts and Terminology

To use `openpyxl`  the basic Excel structure are:

- **Workbook**: The Excel file itself, containing one or more worksheets.
- **Worksheet**: Individual tabs/sheets within a workbook.
- **Cell**: Individual data points in a worksheet, identified by column letter and row number (e.g., `"A1"`).
- **Row**: Horizontal line of cells, identified by numbers (`1`, `2`, `3`, ...).
- **Column**: Vertical line of cells, identified by letters (`A`, `B`, `C`, ...).

In `openpyxl`, you can reference cells using either:

1. **Excel-style references**
    ```python
    sheet["A1"]
    ```

2. **Row-column indexing**
    ```python
    sheet.cell(row=1, column=1)
    ```

> **Note:** `openpyxl` uses **1-based indexing**, not **0-based indexing**.

````python
from openpyxl import Workbook

wb=Workbook()
sheet=wb.active
sheet.title="student Report"

# Title

sheet["A1"]="student Report card"


#headers
headers=[
    "ID",
    "name",
    "Age",
    "Math",
    "Science",
    "English",
    "Average",
    "Grade"
]

# add headers to row 3
for col_num, header in enumerate(headers,1):
    sheet.cell(row=3,column=col_num,value=header)

students = [
    [1, "Alice", 16, 85, 90, 88, 87.6, "A"],
    [2, "Bob", 17, 78, 82, 80, 80.0, "B"],
    [3, "Charlie", 16, 92, 88, 95, 91.6, "A"],
    [4, "David", 17, 70, 75, 72, 72.3, "C"],
    [5, "Emma", 16, 89, 91, 84, 88.0, "A"]
]

#starting row for data
start_row=4

for row_num,student in enumerate(students,start_row):
    for col_num,value in enumerate(student,1):
        sheet.cell(row=row_num,column=col_num,value=value)

wb.save(filename="student_report.xlsx")
````
