```python
# This code joins names from the list of filenames to the end of a folders name. 

from pathlib import Path
my_files=["accounts.txt","naval.csv","Ravikant.docx"]
for filenames in my_files:
    print(Path(r"C:\Users\thapa.pra",filenames))
```

In windows backslash seperates directories, so we cant use it in filenames. However, we can use backslashes in filenames in MacOS and linux. so it is a good idea to always use forward slashes (/) in the python code. The Pathlib module ensures that my code always works on all operating systems.

## Accessing Home Directory
### Home Directory vs Working Directory in Python

- **Home Directory**: Your personal default folder on the computer where your files are usually stored. It is fixed for your user account.  
  - Example: `/home/username` on Linux or `C:/Users/username` on Windows.

- **Working Directory**: The current folder where your program looks for files or saves files. It can be changed anytime using commands like `os.chdir()`.  
  - By default, it often starts as the home directory but can be different during program execution.

**Key Point:**  
The home directory is your starting point, while the working directory is where your program operates at a given moment. Changing the working directory affects file operations relative to that new folder.

```python
from pathlib import Path
Path.home()
```

The home directories are located in set place depending on OS:  
- On windows: `C:\Users`
- On macOS: `/Users`
- On linux: `/home


## Specifying Absolute Path and Relative Path:

- An absolute path is always begins with root folder (C:\on windows and /on macOS and linux)
- A relative path is relative to current working directory. 

    There are also dot `(.)` and dot-dot`(..)` Folders, which are special names can be used in a filepath. A single period(.) is shorthand for this folder. Two periods (..) dot-dot means the parent folder. 
- Absolute Starts with `d:/, C:/, or Path.home()`
- RelativeStarts with `folder name, ./, ../, etc.`

### Getting the Parts of a filepath:

- `Anchor`: Root folder of the filesystem, on windows which is single letter denotes Hard-drive or other storage devive.
- `parent`: is the folder that contains the file.
- `name` of the file, made up of `stem(or base name)` and the `suffix (or extension)`.
